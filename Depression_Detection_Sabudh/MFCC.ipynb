```python
import numpy as np
from librosa import display
```

```python
#Loop to generate all the spectrograms
for i in range (301,345):
  audio_file=r"/content/drive/MyDrive/Depression_Detection/Healthy_People_Audio/AIC_"+str(i)+"_result.mp3"
  y,sr=librosa.load(audio_file)
  mfcc_features=librosa.feature.mfcc(y=y,sr=sr)
  import matplotlib.pyplot as plt
  fig, ax = plt.subplots()
  img = display.specshow(mfcc_features, x_axis='time', ax=ax)
  fig.colorbar(img, ax=ax)
  ax.set(title='MFCC')
  savename=r"/content/drive/MyDrive/Depression_Detection/Healthy_People_Audio_MFCC/MFCC_AIC_"+str(i)+"_result.png"
  fig.savefig(savename)  
  plt.close('all')
```
