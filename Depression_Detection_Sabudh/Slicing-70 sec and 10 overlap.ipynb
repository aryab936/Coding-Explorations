```python
from mutagen.mp3 import MP3
for i in range(301,345):
  filename=r"/content/drive/MyDrive/Depression_Detection/Without_Slicing/Healthy_People_Audio/AIC_"+str(i)+"_result.mp3"
  audio=MP3(filename)
  audio_info=audio.info
  length_in_seconds=int(audio_info.length)
  #print(length_in_seconds)
  audio=AudioSegment.from_mp3(filename)
# 70 sec with 10 overlap
  j=0
  for j in list(range(0,length_in_seconds*1000-80000,70000)):
    audio_crop=audio[j:j+80000]
    filename=r"/content/drive/MyDrive/Depression_Detection/Slicing_Experiment/70_sec_and_10_overlap/Audio_Slices_Healthy/AIC_"+str(i)+"_result_slice_"+str(j)+".mp3"
    audio_crop.export(filename,format="mp3")
```
