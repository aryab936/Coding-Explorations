```python
from scipy.io.wavfile import read
import numpy as np
audio="/content/drive/MyDrive/Depression_Detection/glottal_test_laugh.wav"
size_frame=0.2
size_step=0.05
fs, data_audio=read(audio)
data_audio=data_audio.reshape(-1)
data_audio=data_audio-np.mean(data_audio)
data_audio=data_audio/float(np.max(np.abs(data_audio)))

size_frameS=size_frame*float(fs)
size_stepS=size_step*float(fs)
overlap=size_stepS/size_frameS
```

```python
size_frameS
```

```python
nF=int((len(data_audio)/size_frameS/overlap))-1
data_audiof=np.asarray(data_audio*(2**15), dtype=np.float32)
```

```python
data_audiof.shape
```

```python
import pysptk
```

```python
f0=pysptk.sptk.rapt(data_audiof, fs, int(0.01*fs), min=20, max=500, voice_bias=0, otype='f0')
```
