```python
import numpy as np
from librosa import display
```

```python
for i in range (301,304):
  j=0
  while(True):
    try:
      audio_file=r"/content/drive/MyDrive/Depression_Detection/With_Slicing/Healthy_People_Audio_sliced_train/AIC_"+str(i)+"_result_slice_"+str(j)+".mp3"
      y,sr=librosa.load(audio_file)
      mfcc_features=librosa.feature.mfcc(y=y,sr=sr)
      import matplotlib.pyplot as plt
      fig, ax = plt.subplots()
      img = display.specshow(mfcc_features, x_axis='time', ax=ax)
      fig.colorbar(img, ax=ax)
      ax.set(title='MFCC')
      savename=r"/content/drive/MyDrive/Depression_Detection/With_Slicing/Healthy_People_Audio_sliced_MFCC_train/MFCC_AIC_"+str(i)+"_result_slice_"+str(j)+".png"
      fig.savefig(savename)  
      plt.close('all')
      j+=50000
    except:
      break 
```
